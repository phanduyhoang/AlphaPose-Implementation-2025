```python
import cv2
import numpy as np
import os

# Paths
video_path = "/content/0205.mp4"
weights_path = "/content/yolov3.weights"
cfg_path = "/content/yolov3.cfg"
names_path = "/content/coco.names"
output_video_path = "/content/output_yolo3.mp4"

# Download YOLOv3 files if not available
if not os.path.exists(weights_path):
    !wget -q https://pjreddie.com/media/files/yolov3.weights -O {weights_path}
if not os.path.exists(cfg_path):
    !wget -q https://github.com/pjreddie/darknet/blob/master/cfg/yolov3.cfg?raw=true -O {cfg_path}
if not os.path.exists(names_path):
    !wget -q https://github.com/pjreddie/darknet/blob/master/data/coco.names?raw=true -O {names_path}

# Load YOLO model
net = cv2.dnn.readNet(weights_path, cfg_path)
net.setPreferableBackend(cv2.dnn.DNN_BACKEND_OPENCV)
net.setPreferableTarget(cv2.dnn.DNN_TARGET_CPU)  # Change to CUDA if needed

# Load class labels
with open(names_path, "r") as f:
    classes = [line.strip() for line in f.readlines()]

# Get YOLO output layers
layer_names = net.getLayerNames()
output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers().flatten()]

# Load video
cap = cv2.VideoCapture(video_path)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define video writer
out = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (frame_width, frame_height))

# Process video frames
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    height, width = frame.shape[:2]

    # Convert frame to YOLO input format
    blob = cv2.dnn.blobFromImage(frame, 1/255.0, (416, 416), swapRB=True, crop=False)
    net.setInput(blob)

    # Forward pass
    detections = net.forward(output_layers)

    # Draw bounding boxes
    for output in detections:
        for detection in output:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.3:  # Draw only high-confidence detections
                center_x, center_y, w, h = (detection[:4] * np.array([width, height, width, height])).astype("int")
                x, y = int(center_x - w / 2), int(center_y - h / 2)
                color = (0, 255, 0)  # Green box
                cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)
                label = f"{classes[class_id]}: {confidence:.2f}"
                cv2.putText(frame, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    out.write(frame)  # Save frame to output video

cap.release()
out.release()

print(f"✅ Detection finished! Video saved at {output_video_path}")

# Display output video
from IPython.display import display, HTML
display(HTML(f'<video width="640" height="360" controls><source src="{output_video_path}" type="video/mp4"></video>'))
```
